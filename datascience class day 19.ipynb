```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('sample_pivot.csv')
df.head()
```

```python
df.shape
```

```python
# checking for missing values

df.isnull().sum()
```

```python
# filling the missing values 
```

```python
# forward fill
df1= df.copy()
df1['Units'].fillna(method ='ffill', inplace = True)
df1
```

```python
# forward fill
df1= df.copy()
df1['Units'].fillna(method ='bfill', inplace = True)
df1
```

```python
# using a statistical method to fill
df1= df.copy()
df1['Units'].fillna(df1['Units'].median(),inplace = True)
df1
```

```python
# replacement method to fill
df1= df.copy()
df1['Units'].replace({np.nan:0},inplace = True)
df1
```
