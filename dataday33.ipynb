```python
import numpy as np
import pandas as pd
import matplotlib as sns
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('house_data.csv')
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence'], axis=1, inplace=True)
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
categorical = df.select_dtypes(include=['object', 'category'])
numerical = df.select_dtypes(include=['int', 'float'])
for x in categorical:
    df[x].fillna(df[x].mode()[0], inplace=True)

for x in numerical:
    df[x].fillna(df[x].mean(), inplace=True)

df.isnull().sum().sort_values(ascending=False)
```

```python
df.drop('Id', axis =1 , inplace =True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
categorical = df.select_dtypes(include=['object', 'category'])
encoder=LabelEncoder()
for x in categorical:
    df[x]=encoder.fit_transform(df[x])
```

```python
df.head()
```

```python
x=df.drop('SalePrice', axis=1)
y=df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error,r2_score
pred1=model1.predict(xtest)
print(mean_absolute_error(ytest,pred1))
print(mean_absolute_percentage_error(ytest,pred1))
print(mean_squared_error(ytest,pred1))
print(root_mean_squared_error(ytest,pred1))
print(r2_score(ytest,pred1))
```

```python
from sklearn.tree import DecisionTreeRegressor
model2= DecisionTreeRegressor()
model2.fit(xtrain, ytrain)
```

```python
pred2=model2.predict(xtest)
print(mean_absolute_error(ytest,pred2))
print(mean_absolute_percentage_error(ytest,pred2))
print(mean_squared_error(ytest,pred2))
print(root_mean_squared_error(ytest,pred2))
print(r2_score(ytest,pred2))
```

```python
from sklearn.ensemble import RandomForestRegressor
model3=RandomForestRegressor()
model3.fit(xtrain,ytrain)
```

```python
pred3=model3.predict(xtest)
print(mean_absolute_error(ytest,pred3))
print(mean_absolute_percentage_error(ytest,pred3))
print(mean_squared_error(ytest,pred3))
print(root_mean_squared_error(ytest,pred3))
print(r2_score(ytest,pred3))
```
